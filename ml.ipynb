```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('boston.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_train)
```

```python
model = LinearRegression()
model.fit(X_poly, y_train)
```

```python
poly_transform_data = poly.transform(X_test)
predict_data = model.predict(poly_transform_data)
```

```python
from sklearn.metrics import mean_absolute_error as mae
df = pd.DataFrame({'Tahmin Edilen': predict_data, 'Gerçek Değer': y_test})
print(df)
error = mae(y_test, predict_data)
print("Mean absolute error : " + str(error))
```

```python
plt.plot(predict_data, label='Predicted Values', color='blue')
plt.plot(y_test, label='Actual Values', color='red')
plt.title('Boston House Prices')
```

```python
plt.scatter(X_test[:,6],predict_data, label='Predicted Values', color='blue', marker='x')
plt.scatter(X_test[:,6],y_test, label='Actual Values', color='red')
plt.xlabel('Age Range')
plt.ylabel('Prices')
plt.title('Boston House Prices')
```
